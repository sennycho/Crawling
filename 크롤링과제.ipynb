# 방법1

```python
import requests
from bs4 import BeautifulSoup
```

```python
header = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36'}

data =[]

for i in range(1, 5): 
    genie = f'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20230601&hh=15&rtm=Y&pg={i}'
    request = requests.get(genie, headers=header)
    soup = BeautifulSoup(request.text)
    table = soup.findAll('td', {'class':'info'})
    for j in range(len(table)):
        titles = table[j].find('a', {'class': 'title ellipsis'}).text.strip()
        artists = table[j].find('a', {'class': 'artist ellipsis'}).text.strip()
        print('{0:3d}위 {1} - {2}'.format(50*(i-1)+ j+1, artists, titles))
        dic = {'순위': 50*(i-1)+ j+1, 'title':titles, 'artist':artists}
        data.append(dic)

```

```python
import pandas as pd

df = pd.DataFrame(data)
df
```

```python
df.to_excel('genie.xlsx')
```

# 방법2

```python
from pandas.io.formats.style_render import Index
header = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36'}
num = ['1', '2', '3', '4']
 
data = []

index = 1

for i in num: 
    genie = f'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20230601&hh=15&rtm=Y&pg={i}'
    request = requests.get(genie, headers=header)
    soup = BeautifulSoup(request.text)
    td = soup.findAll('td', {'class':'info'})
    titles = [item.find('a', {'class': 'title ellipsis'}) for item in td]
    artists = [item.find('a', {'class': 'artist ellipsis'}) for item in td] 


    for t,a in zip(titles, artists):
        
        title = t.text.strip()
        artist = a.text.strip()
        print('{0:3d}위 {1} - {2}'.format(index, artist, title))

        dic = {'rank': index ,'title':title, 'artist':artist}
        data.append(dic)

        index += 1


```

```python
df = pd.DataFrame(data)
df
```

```python
df.to_excel('genie2.xlsx')
```
